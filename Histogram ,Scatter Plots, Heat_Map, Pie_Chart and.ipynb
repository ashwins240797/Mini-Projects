```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_boston
from matplotlib import style
%matplotlib inline

boston_dataset = load_boston()
print(boston_dataset.DESCR)
```

```python
x_axis = boston_dataset.data
y_axis = boston_dataset.target
```

```python
# plot histogram
style.use('ggplot')
plt.figure(figsize=(9,9))
plt.hist(y_axis,bins=50)
plt.xlabel('number of houses')
plt.ylabel('price in 1000s USD')
plt.show()
```

```python
# plot scatter plots
style.use('ggplot')
plt.figure(figsize=(7,7))
plt.scatter(boston_dataset.data[:,5],boston_dataset.target,color='b')
plt.xlabel('number of houses')
plt.ylabel('price in 1000s USD')
plt.show()
```

```python
import seaborn as sns

flight_data = sns.load_dataset('flights')
flight_data.head()
```

```python
# use pivot method to re-arrange the datset
flight_data = flight_data.pivot('month','year','passengers')
flight_data
```

```python
# generating heat map
sns.heatmap(flight_data)
```

```python
# PIE CHART representation
job_data = ['40','20','17','8','5','10']

# domains
domain = 'IT','Marketing','Finance','Automotive','Airlines','Healthcare'

# explode the first slice
ex = (0.05,0,0,0,0,0)

#plot the pie chart
plt.pie(job_data,labels=domain,explode=ex,autopct='%1.1f%%')
plt.show()
```
