```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

from PIL import Image

import warnings
warnings.filterwarnings('ignore')
```

```python
# loading the image
img = Image.open("D://M_AI//Machine Learning//Projects//dogs.jpeg")
```

```python
plt.imshow(img)
```

```python
# converting the image into array
img_array = np.array(img)
print(img_array)
```

```python
img_array.shape
```

```python
# coverting image to 2d array
img_array = img_array.reshape((185*272,3))
img_array.shape
```

```python
# perform Kmeans clustering for k = 3
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
kmeans.fit(img_array)
```

```python
# predict the labels
labels = kmeans.predict(img_array)
labels.shape
```

```python
# ploting back as image
labels = labels.reshape(185,272)
```

```python
plt.imshow(labels)
```

```python
kmeans.cluster_centers_
```

```python
print(kmeans.labels_)
print(len(kmeans.labels_))
```

```python
# finding the dominant colors
centers = [centers.astype(int) for centers in kmeans.cluster_centers_]
blob = np.ones((100,100,3))
```

```python
blob1 = blob * centers[0]
plt.imshow(blob1)
```

```python
blob2 = blob * centers[1]
plt.imshow(blob2)
```

```python
blob3 = blob * centers[2]
plt.imshow(blob3)
```
