```python
import matplotlib.pyplot as plt
from matplotlib import style
%matplotlib inline

# give temparature, wind, humidity, precipitation and time data
temp = [78,75,74,92,91,98]
wind = [14,12,10,13,9,25]
time = [1,2,3,4,5,6]
humidity = [73,76,84,79,71,46]
precepitation = [26,42,69,48,11,16]
```

```python
# subplots
plt.figure(figsize=(9,9))
plt.subplots_adjust(hspace = 0.25)
plt.subplot(2,2,1)
plt.title('Temp (C)')
plt.plot(time,temp,color = 'b',linestyle='-',linewidth=1)
plt.subplot(2,2,2)
plt.title('Humidity (%)')
plt.plot(time,humidity,color = 'r',linestyle='-',linewidth=1)
plt.subplot(2,2,3)
plt.title('Wind(KPH)')
plt.plot(time,wind,color = 'y',linestyle='-',linewidth=1)
plt.subplot(2,2,4)
plt.title('Precepitation(%)')
plt.plot(time,precepitation,color = 'g',linestyle='-',linewidth=1)
plt.show()
```
